A distribui√ß√£o demogr√°fica do v√≠rus da Hepatite C ocorre por todo o globo, sendo que a maior quantidade de pessoas com o v√≠rus encontra-se no Mediterr√¢neo Oriental e Europa, cada um com 12 milh√µes de infectados. ## 2. Ajeitar as colunas

Como visto na tabela acima, temos uma coluna que est√° atrapalhando a nossa an√°lise, que √© a coluna **Unnamed: 0**. Para retirar essa coluna, usaremos a fun√ß√£o **drop()** do **DataFrame**.

```python
# Excluir a coluna desejada
dados.drop(['Unnamed: 0'], axis=1, inplace=True)

# Mostrar o conjunto de dados
dados
```

Como o nome das colunas est√° em ingl√™s e estamos no Brasil, nada melhor do que renomear as colunas para saber o que cada uma carrega de informa√ß√£o. Para isso, usaremos a fun√ß√£o **.rename()** em nosso objeto **DataFrame**.

```python
# Para renomear as colunas do DataFrame, precisamos passar o valor "True" para o par√¢metro "inplace"
# Esse par√¢metro informa ao Python que queremos que a tabela seja editada e sempre que a chamamos seja mostrada
# com os nomes que colocamos
# Para isso, vamos criar primeiramente um dicion√°rio com as chaves sendo os nomes antigos e os valores, os novos
nome_colunas = {'Category': 'Categoria', 'Age': 'Idade', 'Sex': 'Sexo'}

# Editar o conjunto de dados
dados.rename(columns=nome_colunas, inplace=True)

# Mostrar o conjunto de dados
dados
```

Como indicado na p√°gina do Kaggle desse estudo, os valores **1=Hepatitis**, **2=Fibrosis** e **3=Cirrhosis** s√£o considerados todos como hepatite. Logo, precisamos converter todos esses valores em uma √∫nica classe que represente essas tr√™s. Al√©m disso, as classes **0=Blood Donor** e **0s=suspect Blood Donor** s√£o consideradas comente como doador. Para isso, usaremos a fun√ß√£o **replace()** na coluna desejada do **DataFrame**.

```python
# 0 ==  0=Blood Donor e 0s=suspect Blood
# 1 == 1=Hepatitis, 2=Fibrosis e 3=Cirrhosis
dados['Categoria'].replace('0=Blood Donor', 0, inplace=True)
dados['Categoria'].replace('0s=suspect Blood Donor', 0, inplace=True)
dados['Categoria'].replace('1=Hepatitis', 1, inplace=True)
dados['Categoria'].replace('2=Fibrosis', 1, inplace=True)
dados['Categoria'].replace('3=Cirrhosis', 1, inplace=True)

# Mostrar a coluna "Categoria"
dados['Categoria']
```

Com a coluna **Categoria** devidamente ajustada, vejamos como os dados est√£o dispostos. Imagine se na coluna **Idade** o *dtype* seja **object** (string) e n√£o **int64** (n√∫mero). Caso isso ocorra com alguma coluna, devemos trocar o seu *dtype* para o tipo correto de informa√ß√£o que ela carrega. Para isso, usaremos a fun√ß√£o **info()** do **DataFrame**.

```python
dados.info()
```

Como podemos ver, os *dtype* de cada coluna est√° conforme o a informa√ß√£o que ela carrega. Mas se voc√™ olhar atentamente, ver√° que h√° coluna com menos valor do que as linhas totais, como em **ALB** que tem 614 valores ao inv√©s de 615. Essa informa√ß√£o mostrada diz a quantidade de dados que n√£o s√£o vazios (**NaN**) e valores vazios atrapalham a an√°lise. Para retirar esses dados **NaN**, usaremos a fun√ß√£o **dropna()** do objeto **DataFrame**.

```python
# "axis=0" para retirar a linha
# "how='any'" para que seja retirada a linha que tenha AO MENOS um valor vazio
# "inplace=True" para que seja editado o DataFrame
dados.dropna(axis=0, how='any', inplace=True)

# Mostrar as informa√ß√µes do conjunto de dados
dados.info()
```

Para deixar os gr√°ficos mais bonitos, vamos chamar duas fun√ß√µes: **set_theme()** do ***seaborn*** e **use()** do ***matplotlib***. A primeira fun√ß√£o serve para que os gr√°ficos matplotlib tenham o mesmo estilo de design dos gr√°ficos seaborn. E a segunda fun√ß√£o serve para determinar o estilo usado nos gr√°ficos.

Agora, vamos definir as cores usadas nos gr√°ficos e algumas fun√ß√µes para que eles sejam mostrados.

```python
# Definir as cores que ser√£o usadas nos gr√°ficos
cores = ['lightcoral', 'deepskyblue', 'orchid', 'tomato', 'teal', 'darkcyan', 'limegreen', 'darkorange']
```

```python
# Definir a fun√ß√£o do gr√°fico de pizza
def grafico_pizza(data_frame, coluna, cores, explode, titulo, fonte):
    # Fazer contagem dos valores da coluna selecionada
    df = data_frame[coluna].value_counts()
    
    # Determinar o tamannho da plotagem
    plt.figure(figsize=(15, 10))
    
    # Criar o gr√°fico de pizza
    _, _, pacotes = plt.pie(df,
                            colors=cores,
                            labels=df.index,
                            explode=explode,
                            shadow=True,
                            startangle=90,
                            autopct='%1.1f%%',
                            textprops={'fontsize': fonte,
                                      'color': 'black',
                                      'weight': 'bold',
                                      'family': 'serif'})
    
    # Plotar o gr√°fico de pizza
    plt.setp(pacotes, color='white')
    
    # Colocar o t√≠tulo do gr√°fico
    plt.title(titulo, size=45)
    
    # Desenhar o c√≠rculo interno
    circulo_centro = plt.Circle((0, 0), 0.40, fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(circulo_centro)
```

```python
# Definir o gr√°fico da fun√ß√£o de distribui√ß√£o
def grafico_distribuicao(data_frame, coluna, titulo):
    # Armazenar os dados da coluna
    dados = data_frame[coluna]
    
    # Determinar a figura e seu tamanho
    fig = plt.figure(figsize=(17, 7))
    
    # Criar a grade em que os gr√°ficos ser√£o plotados
    grade = GridSpec(nrows=2, ncols=1, figure=fig)
    
    # Escolher uma das cores para o gr√°fico
    cor = np.random.choice(cores, 1)[0]
    
    # Motrar o valor de assimetria dos dados
    print(f'Assimetria de {titulo}: {np.round(dados.skew(), 3)}')
    
    # Plotar o histograma
    ax0 = fig.add_subplot(grade[0, :])
    ax0.set_title(f'Histograma e BoxPlot de {titulo}', y=1.05)
    sns.histplot(data=dados, ax=ax0, color=cor)
    
    # Plotar o BoxPlot
    ax1 = fig.add_subplot(grade[1, :])
    plt.axis('off')
    sns.boxplot(x=dados, ax=ax1, color=cor)
```

```python
# Definir o gr√°fico de influ√™ncia
def grafico_influencia(data_frame, coluna, bins, labels, titulo, com_bins=True):
    # Armazenar os dados da coluna
    influencia = data_frame.loc[:, [coluna, 'Categoria']]
    
    # Se os dados de "coluna" n√£o forem como classe, ent√£o ter√° intervalos ("bins" e "labels")
    if com_bins:
        influencia[coluna] = pd.cut(influencia[coluna],
                                   bins=bins,
                                   labels=labels)
        
    # Escolher uma das cores para o gr√°fico
    cor = np.random.choice(cores, 1)[0]
    
    # Determinar o tamanho da figura
    plt.figure(figsize=(15, 5))
    
    # Criar o gr√°fico
    grafico = sns.pointplot(x=coluna, y='Categoria', dodge=0.1, capsize=.1, data=influencia, color=cor)
    
    # Colocar o t√≠tulo do gr√°fico
    grafico.set_title(f'{titulo} influ√™ncia', fontsize=25)
```

## 7. Modelagem do Aprendizado de M√°quina (ou Intelig√™ncia Artificial)

Nesta etapa precisamos separar os dados em dois grupos: o primeiro sendo os dados com todas as informa√ß√µes dos pacientes e o segundo, somente a coluna com o resultado; que no caso √© a coluna **Categoria**, uma vez que √© ela o resultado final que queremos (n√£o adianta entregar a resposta junto com a pergunta n√©?). Para isso, separaremos em duas vari√°veis: o primeiro grupo sendo o **X** e o segundo sendo o **y**. Usaremos essas nota√ß√µes j√° que √© algo padr√£o usado no Aprendizado de M√°quina.

```python
# Separar os dados em dois grupos
X = dados.drop(['Categoria'], axis=1)
y = dados['Categoria']
```

Agora vamos separar os dados em *dados de treino* e *dados de teste*. Separ√°-los serve para que a Intelig√™ncia Artificial (IA) n√£o tenha um overfit. Esse termo √© usado quando uma IA se sai muito bem com os dados de treino, mas quando analisa os dados de teste, ou seja, dados que ela nunca viu, ela se sai bem p√©ssima. √â a mesma coisa quando voc√™ faz uma prova, voc√™ estuda antes com os exerc√≠cios para que se saia bem na avalia√ß√£o, j√° que essa avalia√ß√£o possui quest√µes semelhantes as que voc√™ estudou, mas n√£o iguais. Se forem as mesmas quest√µes do estudo e da prova, qual a certeza de que voc√™ realmente aprendeu a mat√©ria? Isso vale para a IA! Ent√£o separamos uma quantidade aleat√≥ria de dados para que a IA treine e se saia bem com os dados de teste, esses que ela nunca viu.

Para que isso seja feito, vamos usar a fun√ß√£o **train_test_split()** do m√≥dulo ***sklearn***.

```python
from sklearn.model_selection import train_test_split
```

```python
# Informar os arrays para a separa√ß√£o, nesse caso X & y
# Informar o tamanho do conjunto de treino, nesse caso √© de 70% dos dados ser√£o para o treino da IA
# Informar o par√¢metro "random_state" como 0 para que seja sempre os mesmos dados a serem usados quando for
# reproduzir novamente o treinamento da IA
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, train_size=0.7, random_state=0)
```

Com os dados separados em treino e teste, vamos usar tr√™s algoritmos de Aprendizado de M√°quina e ver qual deles √© o melhor. S√£o eles, em ordem alfab√©tica:

- Gradient Boosting Classifier;


- kNN (K-Nearest Neighbours) e;


- Random Forest Classifier.

Todos esses algoritmos est√£o presentes no m√≥dulo ***sklearn***.

```python
# Importar os modelos de Aprendizado de M√°quina
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
```

```python
# Gradient Booting Classifier
gbc = GradientBoostingClassifier(learning_rate=0.01, loss='exponential', max_depth=70,
                                max_features=2, n_estimators=500, random_state=0)

# kNN
knn = KNeighborsClassifier(n_neighbors=5, algorithm='kd_tree', weights='uniform', n_jobs=-1)

# Random Forest Classifier
rfc = RandomForestClassifier(n_estimators=500, max_depth=70, max_features=2, random_state=0, n_jobs=-1)
```

Com os algoritmos montados, basta trein√°-los com o m√©todo **fit()**. Os par√™metros usados s√£o o **X_treino** e **y_treino**. As c√©lulas a seguir podem demorar para ficarem prontas e exigem uma certa capacidade do processador, ent√£o aconselho que somente o Notebook Jupyter esteja aberto em seu computador.

```python
# Treinar o Gradient Booting Classifier
gbc.fit(X_treino, y_treino)
```

```python
# Treinar o kNN
knn.fit(X_treino, y_treino)
```

```python
# Treinar o Random Forest Classifier
rfc.fit(X_treino, y_treino)
```

Com os modelos treinados, vejamos como eles se saem com os dados de teste. Lembre-se que esses dados as IAs nunca viram. Vamos analisar a matriz de confus√£o de cada modelo. Essa matriz serve para observar a taxa de acertos e erros da IA. Para ver essa matriz, chame a fun√ß√£o **confusion_matrix** do m√≥dulo ***sklearn***.

Al√©m de analisar a matriz de confus√£o de cada classificador, veremos sua acur√°cia, precis√£o, revoca√ß√£o e pontua√ß√£o F1. Mas que raios √© tudo isso a√≠? Vejamos cada um e qual a sua finalidade.

- **Acur√°cia**: A pergunta que a acur√°cia responde √©: no geral, o qu√£o frequente o classificador est√° correto? A acur√°cia √© definida por:



- **Precis√£o**: A pergunta que a precis√£o responde √©: daqueles que foram classificados como corretos, quantos efetivamente s√£o corretos? A precis√£o √© definida por:



- **Revoca√ß√£o**: A pergunta que a revoca√ß√£o responde √©: quando realmente √© da classe X, o qu√£o frequente √© classificado como X?. A revoca√ß√£o √© definida por:



- **Pontua√ß√£o F1**: Essa m√©trica combina a precis√£o e a revoca√ß√£o para que haja um n√∫mero √∫nico que indique a qualidade geral do modelo. A pontua√ß√£o F1 √© definida por:



Agora que j√° conhecemos quais s√£o as m√©tricas, vamos import√°-las do m√≥dulo ***sklearn***.

```python
# Importar a matriz de confus√£o e as m√©tricas de avalia√ß√£o
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score, recall_score
```

```python
# Para avaliar as m√©tricas e a matriz de confus√£o, vamos criar uma fun√ß√£o para facilitar a nossa vida
def metricas(X_teste, y_teste, classificador, nome_clf):
    # Prever o resultado
    y_pred = classificador.predict(X_teste)
    
    # M√©trica de acur√°cia
    acuracia = accuracy_score(y_teste, y_pred)
    
    # M√©trica de precis√£o
    precisao = precision_score(y_teste, y_pred)
    
    # M√©trica de revoca√ß√£o
    revocacao = recall_score(y_teste, y_pred)
    
    # M√©trica de pontua√ß√£o F1
    f1 = f1_score(y_teste, y_pred)
    
    # Mostrar os valores das m√©tricas
    print(f'''{nome_clf}

Acur√°cia: {acuracia:.3f}

Precis√£o: {precisao:.3f}

Revoca√ß√£o: {revocacao:.3f}

Pontua√ß√£o F1: {f1:.3f}''')
    
    # Criar a matriz de confus√£o
    matriz = confusion_matrix(y_teste, y_pred)
    
    # Criar um DataFrame para aramazenar os dados de "y_pred" e "y_teste"
    df_mc = pd.DataFrame(matriz, index=['Doador', 'Hepatite C'], columns=['Doador', 'Hepatite C'])
    
    # Determinar o tamanho da plotagem
    plt.figure(figsize=(15, 10))
    
    # Criar a plotagem
    mapa_calor = sns.heatmap(data=df_mc,
                            annot=True,
                            cmap='RdPu',
                            fmt='.5g',
                            annot_kws={'size': 20})
    
    # Colocar o t√≠tulo da matriz de confus√£o
    mapa_calor.set_title(f'Matriz de Confus√£o {nome_clf}', fontsize=25, y=1.05)
    
    # Adicionar a legenda nos eixos
    plt.xlabel('Valores Previstos', fontsize=20)
    plt.ylabel('Valores Reais', fontsize=20)
    
    # Mostrar a matriz de confus√£o
    plt.show()
```

Com a fun√ß√£o devidamente criada, basta cham√°-la para cada classificador.

```python
# Mostrar as m√©tricas e matriz de confus√£o do Gradient Boosting Classifier
metricas(X_teste, y_teste, gbc, 'Gradient Boosting Classifier')
```

```python
# Mostrar as m√©tricas e a matriz de confus√£o de kNN
metricas(X_teste, y_teste, knn, 'kNN')
```

```python
# Mostrar as m√©tricas e a matriz de confus√£o de Random Forest Classifier
metricas(X_teste, y_teste, rfc, 'Radom Forest Classifier')
```
