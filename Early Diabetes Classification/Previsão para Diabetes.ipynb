# Previsão para Diabetes

## Contexto

Durante o período de 2019 a 2021, a Federação Internacional de Diabetes (IDF) relatou um aumento em 16% de casos de diabetes ao redor do mundo em pessoas de 20 a 79 anos. A IDF revelou que em 2021, mais de 74 milhões de pessoas foram acometidas por essa doença, totalizando 537 milhões de pessoas. No Brasil, foram 16,8 milhões de novos casos, cerca de 7% da população `[1]`. Em 2021, ao redor do mundo `[2]`:

- 6,7 mihões de pessoas morreram devido a diabetes (1 morte a cada 5 segundos);


- A receita gasta para o tratamento de diabetes foi de USD 966 bilhões de dólares, um aumento de 316% em relação a 15 anos atrás e;


- A estimativa de diabéticos ao redor do globo em 2030 é de 643 milhões e em 2045 é de 783 milhões de pessoas.

No Brasil, 214 mil pessoas entre 20 e 79 anos vieram a óbito devido a esta enfermidade. A diabetes foi responsável por 2,8% dos óbitos em pessoas com menos de 60 anos em 2021 `[1]`.

Diabetes é uma doença crônica causada pelo aumento execessivo de glicose no sangue, podendo prejudicar diversos órgaos (como os rins e os olhos) se não for tratada. As principais causas para o aumento exagerado de glicose no sangue é a falta de atividade física acompanhada de má alimentação, sendo esse estilo de vida a causa mais comum para o desenvolvimento desta enfermidade. Há quatro grupos principais de classificação da diabetes `[3]`:

- **Diabetes tipo 1**: A diabetes é o tipo menos comum e acomete pacientes desde o seu nascimento, uma vez que o tipo 1 é caracterizado pelo ataque do sitema imnuológico nas células produtoras de insulina (células ß) do pâncreas. Deste modo, a produção de insulina (proteína que permite a entrada da glicose nas células) fica prejudicada, aumentando assim os níveis de glicose no sangue `[4]`.


- **Diabetes tipo 2**: É o tipo da enfermidade mais comum na população mundial. Ela ocorre devido a resitência à insulina pelo corpo, normalmente devido a estilo de vida sedentário e maus hábitos alimentares. Diferente da tipo 1, na tipo 2 o pâncreas continua produzindo a insulina, mas o corpo cria uma resistência a essa proteína, forçando as células ß a produzirem mais insulina para atingir a quantidade mínima necessária para que a glicose entre nas células. Com o passar dos anos, as células ß não dão mais conta de produzir insulina em excesso e por isso acabam morrendo, não havendo reposição dessas células pelo pâncreas. Um dos fatores para o aumento da resistência à insulina é a gordura na região abdominal, uma vez que as células de gordura contribuem para essa resistência `[4]`.


- **Diabetes gestacional**: Esse tipo de diabetes ocorre somente em mulheres durante a gestação devido a hormônios que a placenta produz que bloqueiam parcialmente a ação da insulina. Mulheres com histórico de Diabetes tipo 2 na família, se ganharam muito peso durante a gravidez, se estão com sobrepeso antes da gravidez, já tiveram diabetes gestacional, ou filhos anteriores que nasceram com mais de 4 quilos, possuem maiores chances de desenvolver a diabetes gestacional `[4]`.


- **Pré-diabetes**: Esse quadro ocorre quando a quantidade de glicose no sangue está aumentando, mas não o suficiente para ser diagnosticada como diabetes. Exames sanguíneo de glicemia em jejum que estejam no valor entre 100 a 120 mg/dL de glicose, é considerado um quadro de pré-diabetes `[4]`.

Os sintomas para a **Diabetes tipo 1** são `[5]`:


- Vontade frequente de urinar, mesmo durante a noite;


- Sensação de sede excessiva;


- Fome em excesso;


- Perda de peso sem razão aparente;


- Cansaço frequente;


- Sonolência injustificável;


- Coceira em todo o corpo;


- Infecções frequentes, como candidíase ou infecção urinária e;


- Irritabilidade e mudanças de humor repentinas.


Os sintomas para a **Diabetes tipo 2** são `[5]`:


- Aumento da sede;


- Boca constantemente seca;


- Vontade de urinar;


- Cansaço frequente;


- Visão turva ou embaçada;


- Feridas que cicatrizam lentamente;


- Formigamento nos pés ou nas mãos e; 


- Infecções frequentes, como candidíase ou infecção urinária.

Para a **Diabetes gestacional**, os sintomas são os mesmos que para a **Diabetes tipo 2**, como o aumento da vontade de urinar, fome e sede excessiva, casos esses que podem ser confundidos com sintomas da gravidez. Para ter certeza se é um quadro de diabetes gestacional, os exames solicitados são o de glicemia em jejum e o teste de tolerância a glicose (TTOG) `[5]`.

```python
# Definir as cores que serão usadas nos gráficos
cores = ['lightcoral', 'deepskyblue', 'orchid', 'tomato', 'teal', 'darkcyan', 'limegreen', 'darkorange']
```

```python
# Definir a função do gráfico de pizza
def grafico_pizza(data_frame, coluna, cores, explode, titulo, fonte):
    # Fazer contagem dos valores da coluna selecionada
    df = data_frame[coluna].value_counts()
    
    # Determinar o tamannho da plotagem
    plt.figure(figsize=(15, 10))
    
    # Criar o gráfico de pizza
    _, _, pacotes = plt.pie(df,
                            colors=cores,
                            labels=df.index,
                            explode=explode,
                            shadow=True,
                            startangle=90,
                            autopct='%1.1f%%',
                            textprops={'fontsize': fonte,
                                      'color': 'black',
                                      'weight': 'bold',
                                      'family': 'serif'})
    
    # Plotar o gráfico de pizza
    plt.setp(pacotes, color='white')
    
    # Colocar o título do gráfico
    plt.title(titulo, size=45)
    
    # Desenhar o círculo interno
    circulo_centro = plt.Circle((0, 0), 0.40, fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(circulo_centro)
```

```python
# Definir o gráfico da função de distribuição
def grafico_distribuicao(data_frame, coluna, titulo):
    # Armazenar os dados da coluna
    dados = data_frame[coluna]
    
    # Determinar a figura e seu tamanho
    fig = plt.figure(figsize=(17, 7))
    
    # Criar a grade em que os gráficos serão plotados
    grade = GridSpec(nrows=2, ncols=1, figure=fig)
    
    # Escolher uma das cores para o gráfico
    cor = np.random.choice(cores, 1)[0]
    
    # Motrar o valor de assimetria dos dados
    print(f'Assimetria de {titulo}: {np.round(dados.skew(), 3)}')
    
    # Plotar o histograma
    ax0 = fig.add_subplot(grade[0, :])
    ax0.set_title(f'Histograma e BoxPlot de {titulo}', y=1.05)
    sns.histplot(data=dados, ax=ax0, color=cor)
    
    # Plotar o BoxPlot
    ax1 = fig.add_subplot(grade[1, :])
    plt.axis('off')
    sns.boxplot(x=dados, ax=ax1, color=cor)
```

```python
# Definir o gráfico de influência
def grafico_influencia(data_frame, coluna, bins, labels, titulo, com_bins=True):
    # Armazenar os dados da coluna
    influencia = data_frame.loc[:, [coluna, 'Diabético']]
    
    # Se os dados de "coluna" não forem como classe, então terá intervalos ("bins" e "labels")
    if com_bins:
        influencia[coluna] = pd.cut(influencia[coluna],
                                   bins=bins,
                                   labels=labels)
        
    # Escolher uma das cores para o gráfico
    cor = np.random.choice(cores, 1)[0]
    
    # Determinar o tamanho da figura
    plt.figure(figsize=(15, 5))
    
    # Criar o gráfico
    grafico = sns.pointplot(x=coluna, y='Diabético', dodge=0.1, capsize=.1, data=influencia, color=cor)
    
    # Colocar o título do gráfico
    grafico.set_title(f'{titulo} influência', fontsize=25)
```

## 7. Modelagem do Aprendizado de Máquina (ou Inteligência Artificial)

Nesta etapa precisamos separar os dados em dois grupos: o primeiro sendo os dados com todas as informações dos pacientes e o segundo, somente a coluna com o resultado; que no caso é a coluna **Diabético**, uma vez que é ela o resultado final que queremos (não adianta entregar a resposta junto com a pergunta né?). Para isso, separaremos em duas variáveis: o primeiro grupo sendo o X e o segundo sendo o y. Usaremos essas notações já que é algo padrão usado no Aprendizado de Máquina.

```python
# Separar os dados em dois grupos
X = dados.drop(['Diabético'], axis=1)
y = dados['Diabético']
```

Agora vamos separar os dados em dados de treino e dados de teste. Separá-los serve para que a Inteligência Artificial (IA) não tenha um overfit. Esse termo é usado quando uma IA se sai muito bem com os dados de treino, mas quando analisa os dados de teste, ou seja, dados que ela nunca viu, ela se sai bem péssima. É a mesma coisa quando você faz uma prova, você estuda antes com os exercícios para que se saia bem na avaliação, já que essa avaliação possui questões semelhantes as que você estdou, mas não iguais. Se forem as mesmas questões do estudo e da prova, qual a certeza de que você realmente aprendeu a matéria? Isso vale para a IA! Então separamos uma quantidade aleatória de dados para que a IA treine e se saia bem com os dados de teste, esses que ela nunca viu.

Para que isso seja feito, vamos usar a função **train_test_split()** do módulo ***sklearn***.

```python
from sklearn.model_selection import train_test_split
```

```python
# Informar os arrays para a separação, nesse caso X & y
# Informar o tamanho do conjunto de treino, nesse caso é de 70% dos dados serão para o treino da IA
# Informar o parâmetro "random_state" como 0 para que seja sempre os mesmos dados a serem usados quando for
# reproduzir novamente o treinamento da IA
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, train_size=0.7, random_state=0)
```

Com os modelos treinados, vejamos como eles se saem com os dados de teste. Lembre-se que esses dados as IAs nunca viram. Vamos analisar a matriz de confusão de cada modelo. Essa matriz serve para observar a taxa de acertos e erros da IA. Para ver essa matriz, chame a função **confusion_matrix** do módulo ***sklearn***.

Além de analisar a matriz de confusão de cada classificador, veremos sua acurácia, precisão, revocação e pontuação F1. Mas que raios é tudo isso aí? Vejamos cada um e qual a sua finalidade.

- **Acurácia**: A pergunta que a acurácia responde é: no geral, o quão frequente o classificador está correto? A acurácia é definida por:



- **Precisão**: A pergunta que a precisão responde é: daqueles que foram classificados como corretos, quantos efetivamente são corretos? A precisão é definida por:



- **Revocação**: A pergunta que a revocação responde é: quando realmente é da classe X, o quão frequente é classificado como X?. A revocação é definida por:



- **Pontuação F1**: Essa métrica combina a precisão e a revocação para que haja um número único que indique a qualidade geral do modelo. A pontuação F1 é definida por:



Agora que já conhecemos quais são as métricas, vamos importá-las do módulo ***sklearn***.

```python
# Importar a matriz de confusão e as métricas de avaliação
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score, recall_score
```

Vamos agora criar uma função que une as métricas e a matriz de confusão. Criar essa função ajudará e muito, precisando escrever o código somente uma vez.

```python
# Para avaliar as métricas e a matriz de confusão, vamos criar uma função para facilitar a nossa vida
def metricas(X_teste, y_teste, classificador, nome_clf):
    # Prever o resultado
    y_pred = classificador.predict(X_teste)
    
    # Métrica de acurácia
    acuracia = accuracy_score(y_teste, y_pred)
    
    # Métrica de precisão
    precisao = precision_score(y_teste, y_pred)
    
    # Métrica de revocação
    revocacao = recall_score(y_teste, y_pred)
    
    # Métrica de pontuação F1
    f1 = f1_score(y_teste, y_pred)
    
    # Mostrar os valores das métricas
    print(f'''{nome_clf}

Acurácia: {acuracia:.3f}

Precisão: {precisao:.3f}

Revocação: {revocacao:.3f}

Pontuação F1: {f1:.3f}''')
    
    # Criar a matriz de confusão
    matriz = confusion_matrix(y_teste, y_pred)
    
    # Criar um DataFrame para aramazenar os dados de "y_pred" e "y_teste"
    df_mc = pd.DataFrame(matriz, index=['Não diabético', 'Diabético'], columns=['Não diabético', 'Diabético'])
    
    # Determinar o tamanho da plotagem
    plt.figure(figsize=(15, 10))
    
    # Criar a plotagem
    mapa_calor = sns.heatmap(data=df_mc,
                            annot=True,
                            cmap='RdPu',
                            fmt='.5g',
                            annot_kws={'size': 20})
    
    # Colocar o título da matriz de confusão
    mapa_calor.set_title(f'Matriz de Confusão {nome_clf}', fontsize=25, y=1.05)
    
    # Adicionar a legenda nos eixos
    plt.xlabel('Valores Previstos', fontsize=20)
    plt.ylabel('Valores Reais', fontsize=20)
    
    # Mostrar a matriz de confusão
    plt.show()
```

Com a função devidamente criada, basta chamá-la para cada classificador.

```python
# Mostrar as métricas e matriz de confusão do Gradient Boosting Classifier
metricas(X_teste, y_teste, gbc, 'Gradient Boosting Classifier')
```

```python
# Mostrar as métricas e a matriz de confusão de kNN
metricas(X_teste, y_teste, knn, 'kNN')
```

```python
# Mostrar as métricas e a matriz de confusão de Random Forest Classifier
metricas(X_teste, y_teste, rfc, 'Radom Forest Classifier')
```
